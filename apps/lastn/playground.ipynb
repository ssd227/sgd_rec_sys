```python
import torch
import matplotlib.pyplot as plt

# 创建行和列的向量
rows = torch.arange(0, 5)  # 行坐标
cols = torch.arange(0, 5)  # 列坐标

# 计算外积，得到特征图
feature_map = torch.einsum('i,j->ij', rows, cols)

# 可视化特征图
plt.imshow(feature_map, cmap='viridis')
plt.colorbar()
plt.xlabel('Column')
plt.ylabel('Row')
plt.title('Feature Map')
plt.show()
```
